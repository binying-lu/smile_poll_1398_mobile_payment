```python
import pandas as pd
import datetime
from scipy import stats
```

```python
df=pd.read_csv("/Users/binying.lu/Documents/GitHub/smile_poll_1398_mobile_payment/1398.csv")
df
```

```python
pd_df=pd.DataFrame()
pd_df["Gender"]=df["Gender"].apply(lambda x: "男" if x==1 else "女")
pd_df["Born year"]=pd.to_datetime(df["Born year"]).dt.year
pd_df["age"]=2025-pd_df["Born year"].astype(int)
pd_df
```

```python
def age_range(age):
    return f"{int(age/10)*10+1}-{(int(age/10)*10)+10}歲"
pd_df["age_range"]=pd_df["age"].apply(age_range)

def edu(edu):
    if edu==1:
        return "國小（或以下）"
    elif edu==2:
        return "初中、國中"
    elif edu==3:
        return "高中、高職"
    elif edu==4:
        return "專科"
    elif edu==5:
        return "大學"
    elif edu==6:
        return "研究所（或以上）"
pd_df["edu"]=df["EDU"].apply(edu)
pd_df["edu_num"]=df["EDU"]

def convenient(values):
    if values==1 or values==2:
        return 1
    elif values==3 or values==4:
        return 0
    else:
        return None
pd_df["convenient"]=df["你認為使用行動支付是更為方便還是更為麻煩？"].apply(convenient)

def safety(values):
    if values==1 or values==2:
        return 1
    elif values==3 or values==4:
        return 0
    else:
        return None
pd_df["safety"]=df["你認為使用行動支付是更為安全還是更為危險？"].apply(safety)

pd_df["using"]=df["你有使用過行動支付嗎？"].apply(lambda x:1 if x==1 else 0)

pd_df=pd_df.dropna()
pd_df
```

```python
r,p=stats.pearsonr(pd_df["using"],pd_df["convenient"])
r_squared =r**2
print(f"Pearson correlation={r_squared}")
print(f"p-value={p}")
```

```python
r,p =stats.pearsonr(pd_df["using"],pd_df["safety"])
r_squared =r**2
print(f"Pearson correlation={r_squared}")
print(f"p-value={p}")
```
